```python
import datetime as dt
import pyperclip
import webbrowser

entry = pyperclip.paste()

entry = entry.replace('\n'," ").replace('\r',' ').split("  ")[4:]

def split_by_empty_string(data):
    result = []
    current_list = []
    
    for item in data:
        if item == '':
            if current_list:
                result.append(current_list)
                current_list = []
        else:
            current_list.append(item)
    
    if current_list:
        result.append(current_list)
    return result

def join_lists_with_comma(lists):
    return [' '.join(sublist) for sublist in lists]

split_lists = split_by_empty_string(entry)
joined_lists = join_lists_with_comma(split_lists)

def ile_do_końca(entry):
    entry_splited = entry.split(" ")
    if len(entry_splited) == 15:
        card_id = entry_splited[0]
        card_name = " ".join([entry_splited[1],entry_splited[2]])
        entry_time = dt.datetime.strptime(str(entry_splited[5]+" "+entry_splited[6]), "%d.%m.%Y %H:%M")
        end_time = entry_splited[13]
        minuty, sekundy = map(int, end_time.split(','))
        czas_delta = dt.timedelta(minutes=minuty, seconds=sekundy)
        dni = czas_delta.days
        godziny, reszta = divmod(czas_delta.seconds, 3600)
        minuty = reszta // 60
        wyprawa=entry_splited[8]
        
        print('_'*70)
        print(f"{card_name} {card_id} Czas jaki spędzi na {wyprawa} wyprawie to\n| {dni} dni | {godziny} godzin | {minuty} minut |\n")
        outcome = entry_time+czas_delta
        print('Wyprawa zakończy się:\n', outcome)

        delta = outcome-dt.datetime.now()
        formatted_delta = "| {:d} dni | {:02} godzin | {:02} minut | {:02} sekund |".format(delta.days, delta.seconds // 3600, (delta.seconds // 60) % 60, delta.seconds % 60)
        print("\nPozostało:", formatted_delta)
        output={'card_id':card_id,'card_name':card_name,'entry_time':entry_time,'end_time':end_time,'minuty':minuty,'sekundy':sekundy,'czas_delta':czas_delta,"dni":dni,'godziny':godziny,"wyprawa":wyprawa}
        return output
    elif len(entry_splited) == 16:
        card_id = entry_splited[0]
        card_name = " ".join([entry_splited[1],entry_splited[2],entry_splited[3]])
        entry_time = dt.datetime.strptime(str(entry_splited[6]+" "+entry_splited[7]), "%d.%m.%Y %H:%M")
        end_time = entry_splited[14]
        minuty, sekundy = map(int, end_time.split(','))
        czas_delta = dt.timedelta(minutes=minuty, seconds=sekundy)
        dni = czas_delta.days
        godziny, reszta = divmod(czas_delta.seconds, 3600)
        minuty = reszta // 60
        wyprawa=entry_splited[9]
        
        print('_'*70)
        print(f"{card_name} {card_id} Czas jaki spędzi na {wyprawa} wyprawie to\n| {dni} dni | {godziny} godzin | {minuty} minut |\n")
        outcome = entry_time+czas_delta
        print('Wyprawa zakończy się:\n', outcome)

        delta = outcome-dt.datetime.now()
        formatted_delta = "| {:d} dni | {:02} godzin | {:02} minut | {:02} sekund |".format(delta.days, delta.seconds // 3600, (delta.seconds // 60) % 60, delta.seconds % 60)
        print("\nPozostało:", formatted_delta)
        output={'card_id':card_id,'card_name':card_name,'entry_time':entry_time,'end_time':end_time,'minuty':minuty,'sekundy':sekundy,'czas_delta':czas_delta,"dni":dni,'godziny':godziny,"wyprawa":wyprawa}
        return output
    else:
        card_id = entry_splited[0]
        card_name = entry_splited[1]
        entry_time = dt.datetime.strptime(str(entry_splited[4]+" "+entry_splited[5]), "%d.%m.%Y %H:%M")
        end_time = entry_splited[12]
        minuty, sekundy = map(int, end_time.split(','))
        czas_delta = dt.timedelta(minutes=minuty, seconds=sekundy)
        dni = czas_delta.days
        godziny, reszta = divmod(czas_delta.seconds, 3600)
        minuty = reszta // 60
        wyprawa = entry_splited[7]
        
        print('_'*70)
        print(f"{card_name} {card_id} Czas jaki spędzi na {wyprawa} wyprawie to\n| {dni} dni | {godziny} godzin | {minuty} minut |\n")
        outcome = entry_time+czas_delta
        print('Wyprawa zakończy się:\n', outcome)

        delta = outcome-dt.datetime.now()
        formatted_delta = "| {:d} dni | {:02} godzin | {:02} minut | {:02} sekund |".format(delta.days, delta.seconds // 3600, (delta.seconds // 60) % 60, delta.seconds % 60)
        print("\nPozostało:", formatted_delta)
        output={'card_id':card_id,'card_name':card_name,'entry_time':entry_time,'end_time':end_time,'minuty':minuty,'sekundy':sekundy,'czas_delta':czas_delta,"dni":dni,'godziny':godziny,"wyprawa":wyprawa}
        return output

with open('czas wyprawy.txt', mode='w', encoding='utf8') as nowy_plik:
    for i in joined_lists:
        a = ile_do_końca(i)
        line = '_'*70+'\n'
        nowy_plik.write(line)
        nowy_plik.write(str(a['card_name'])+' '+a['card_id']+' Czas jaki spędzi na '+a['wyprawa']+' wyprawie to\n| '+ str(a['dni']) +' dni | '+str(a['godziny'])+' godzin | '+str(a['minuty'])+' minut |\n')
        outcome=a['entry_time']+a['czas_delta']
        nowy_plik.write('Wyprawa zakończy się:\n '+str(outcome))
        delta = outcome-dt.datetime.now()
        formatted_delta = "| {:d} dni | {:02} godzin | {:02} minut | {:02} sekund |".format(delta.days, delta.seconds // 3600, (delta.seconds // 60) % 60, delta.seconds % 60)
        nowy_plik.write('\nPozostało: '+str(formatted_delta)+'\n')
        
nowy_plik.close()
webbrowser.open('czas wyprawy.txt')
```
